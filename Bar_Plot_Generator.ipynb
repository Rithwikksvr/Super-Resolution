```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
models = ['Linear','Ridge','KNR(Uniform)','KNR(Inverse)','Matrix']
train_psnr = [100,54,20,100,23.71]

plt.bar(np.arange(0,len(models))*2,train_psnr,.5)
plt.xticks((np.arange(0,len(models))*2)+.5,models)
plt.title('PSNR Scores on Training Data for Different Algorithms')
plt.show()
```

```python
train_errors = [3e-31,4.2823610843435156e-06,0.011260383100201748,0,0.0046642396059277417]

plt.bar(np.arange(0,len(models))*2,train_errors,.5)
plt.xticks((np.arange(0,len(models))*2)+.5,models)
plt.title('MSE Errors on Training Data for Different Algorithms')
plt.show()
```

```python
test_mse = [0.012806235901562874,0.012806021479384947,0.043984483535067453,0.043881272658752517,0.0053622029594153639]
plt.bar(np.arange(0,len(models))*2,test_mse,.5)
plt.xticks((np.arange(0,len(models))*2)+.5,models)
plt.title('MSE Errors on Test Data for Different Algorithms')
plt.show()
```
