```python
import cv2
import numpy
import math
import os
import sklearn.linear_model
import sklearn.datasets
import matplotlib.pyplot as plt
import autograd.numpy as np
from autograd import grad
plt.close('all')
```

```python
def cost(A,B,X,y):
    pred = np.matmul(np.matmul(A,X),B)
    return np.mean(np.square(pred - y))
    
```

```python
img = cv2.imread('C:/Users/Rithwik/Desktop/Project Course/faces/img_0002.jpg')
```

```python
folder = 'faces'
lst = os.listdir(folder)
hr_img = cv2.resize(cv2.imread('faces/'+lst[0],0),(0,0),fx=0.5,fy=.5)/255
lr_img = hr_img[::2,::2]
(h,w) = hr_img.shape
```

```python
hr_img.shape
```

```python
lr_img.shape
```

```python
A = np.random.random((h,h//2))/10
B = np.random.random((w//2,w))/10
```

```python
A.shape
```

```python
pred = np.matmul(np.matmul(A,lr_img),B)

cost(A,B,lr_img,hr_img)

```

```python
trainer = grad(cost,argnum=[0,1])
alpha = .001
iterations = 10000


for i in range(iterations):
    grad_A,grad_B = trainer(A,B,lr_img,hr_img)
    A -=alpha*grad_A
    B -=alpha*grad_B
    if i%100==0:
        print (i,cost(A,B,lr_img,hr_img))
    
    
    
    
```

```python
res = np.matmul(np.matmul(A,lr_img),B)
```
