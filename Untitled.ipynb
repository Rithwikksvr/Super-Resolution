```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('faces/image_0001.jpg',0)
img = cv2.resize(img,(0,0),fx=.5,fy=.5)*1.0/255
h,w = img.shape
```

```python
img
```

```python
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
k = 3

blurred = cv2.blur(cv2.resize(img,(0,0),fx=.5,fy=.5),(k,k))
plt.close('all')
plt.imshow(blurred,cmap='gray')
plt.title('bLurred')
plt.show()
```

```python
thresh = .99
noise = np.random.random((h//2,w//2))
org_copy = np.copy(img)
org_copy = org_copy[::2,::2]
org_copy[noise>thresh] = 1

noise = np.random.random((h//2,w//2))
org_copy[noise>thresh] = 0
```

```python
plt.imshow(org_copy,cmap='gray')
plt.title('Generated Image')
plt.show()
```
